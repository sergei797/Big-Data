```python
import requests
from bs4 import BeautifulSoup
import time
import json
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time

%matplotlib inline
```

```python
# парсинг
url = 'https://www.kommersant.ru/doc/6311087?from=glavnoe_3'
firstpage = 4148690
lastpage = 4148790
```

```python
# функция перебора страниц
def get_urls():
  listUrl = []
  for page in range(firstpage, lastpage):
    pageUrl = '{} {}'.format(url, page)
    listUrl.append(pageUrl)
  return listUrl
```

```python
get_urls()
```

```python
postList = []
def get_data():
  for pageData in get_urls():
    fullData = requests.get(pageData)
    print(fullData.status_code)

    if fullData.status_code == 200:
      soup = BeautifulSoup(fullData.text, 'html5lib')

      posts = soup.findAll('div', class_ = 'article_text_wrapper')
      for post in range(len(posts)):
        postList.append(posts[post].text.replace('!','').replace('?', '').
                        replace('-', '').replace('.', '').replace("n/","").
                        replace(",","").lower())
  time.sleep(2.0)

  return postList
```

```python
get_data()
```

```python
# создаем dataframe
columns = ['news']
data = postList
df = pd.DataFrame(data, columns = columns)
df.head()
```

```python
# токенизация
import nltk
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
```

```python
# новая колонка стокенами
df['tokens'] = df['news'].apply(tokenizer.tokenize)
df.head()
```

```python
# разделение на слова
df['tokens'][1][10]
```

```python
# словарь все слова в индексы
def createVocabulary(allWords):
  # пустой словарь сключами по всем словам
  wCount = dict.fromkeys(allWords, 0)

  # считаем частоту появления каждого слова
  for word in allWords:
    wCount[word] +=1

  print('Собранная частота появления слов')

  # сортируем лист по количеству появления слов
  wordsList = list(wCount.items())
  wordsList.sort( key = lambda i: i[1], reverse = 1)

  # список слов отсортированных по частоте появления
  sortedWords = []

  for word in wordsList:
    sortedWords.append(word[0])

  # назначаем индесксы
  wordIndexes = dict.fromkeys(allWords, 0)
  wordKeys = wordIndexes.keys()
  wordKeysLen = len(wordKeys)

  index = 0
  lastPer = 0

  for word in wordKeys:
    wordIndexes[word] = sortedWords.index(word) + 1
    index += 1

    per = round(100 * index / wordKeysLen)

    if (((per % 10) == 0) & (lastPer != per)):
      print (per, '% обратно', sep = "")
      lastPer = per

  print('Собран словарь частотности')

  return wordIndexes
```
