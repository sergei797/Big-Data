```python
from sklearn.datasets import make_classification
import numpy as np
import pandas as pd
from matplotlib import pyplot
from collections import Counter
%matplotlib inline
```

```python
# создаем датасет
x, y = make_classification(n_samples = 10000, n_features = 2, n_redundant = 0, n_clusters_per_class = 1, weights = [0.99],
random_state = 1)
```

```python
counter = Counter(y)
```

```python
# визуализации датасета
for label, _ in counter.items():
    row_ix = np.where( y == label)[0]
    pyplot.scatter(x[row_ix, 0], x[row_ix, 1], label = str(label))
    pyplot.legend()
    pyplot.show()
```

```python
import imblearn
```

```python
from imblearn.under_sampling import NearMiss, RandomUnderSampler

```

```python
undersampling_v1 = NearMiss(version = 1)
```

```python
# трансфармируем датасет
x, y = undersampling_v1.fit_resample(x, y)
counter = Counter(y)
print(counter)
# визуализации датасета
for label, _ in counter.items():
    row_ix = np.where( y == label)[0]
    pyplot.scatter(x[row_ix, 0], x[row_ix, 1], label = str(label))
    pyplot.legend()
    pyplot.show()
```

```python
# версия 2
undersampling_v2 = NearMiss(version = 2)
# трансфармируем датасет
x, y = undersampling_v2.fit_resample(x, y)
counter = Counter(y)
print(counter)
# визуализации датасета
for label, _ in counter.items():
    row_ix = np.where( y == label)[0]
    pyplot.scatter(x[row_ix, 0], x[row_ix, 1], label = str(label))
    pyplot.legend()
    pyplot.show()
```

```python
# версия 3
undersampling_v3 = NearMiss(version = 3)
# трансфармируем датасет
x, y = undersampling_v3.fit_resample(x, y)
counter = Counter(y)
print(counter)
# визуализации датасета
for label, _ in counter.items():
    row_ix = np.where( y == label)[0]
    pyplot.scatter(x[row_ix, 0], x[row_ix, 1], label = str(label))
    pyplot.legend()
    pyplot.show()
```

```python
# OverSampling увеличение количества примеров миноретарного класса
# метод RandomOverSamplingom imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import RandomOverSampler
```

```python
rand_oversampling = RandomOverSampler(sampling_strategy = 'minority')
# трансфармируем датасет
x_over, y_over = rand_oversampling.fit_resample(x, y)
counter = Counter(y_over)
print(counter)
# визуализации датасета
for label, _ in counter.items():
    row_ix = np.where( y_over == label)[0]
    pyplot.scatter(x_over[row_ix, 0], x_over[row_ix, 1], label = str(label))
    pyplot.legend()
    pyplot.show()
```

```python
# метод SMOTE
from imblearn.over_sampling import RandomOverSampler, SMOTE
```

```python
smote = SMOTE()
# трансфармируем датасет
x_over, y_over = smote.fit_resample(x, y)
counter = Counter(y_over)
print(counter)
# визуализации датасета
for label, _ in counter.items():
    row_ix = np.where( y_over == label)[0]
    pyplot.scatter(x_over[row_ix, 0], x_over[row_ix, 1], label = str(label))
    pyplot.legend()
    pyplot.show()
```
