```python
# pip install scikit-learn
# https://scikit-learn.org/stable/modules/clustering.html
import sklearn
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
# Задаем массив
X = np.array([[0, 0],
             [1, 0],
             [0, 1],
             [1, 1],
             [2, 1],
             [2, 2],
             [3, 2],
             [6, 6],
             [7, 6],
             [8, 6],
             [6, 7],
             [7, 7],
             [8, 7],
             [8, 8],
             [9, 8],
             [8, 9],
             [9, 9],])
```

```python
# Нанесение массива данных на плоскость
plt.scatter(X[:, 0], X[:, 1])
```

```python
# Применим метод k-средних
k_means = KMeans(n_clusters = 2)
```

```python
k_means.fit(X)
```

```python
# Печать центров кластеров
k_meansntr = k_means.cluster_centers_
print(k_means.cluster_centers_)
```

```python
# Вывод меток классов кластеров
print(k_means.labels_)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = k_means.labels_ , marker = '^', cmap = 'rainbow')
plt.scatter(k_means.cluster_centers_[:, 1], k_means.cluster_centers_[:, 1], color = 'black' )
```

```python
from sklearn import datasets
```

```python
# скачивание данных с ирисами Фишера
    # 1. Длина наружноц доли околоцветника(sepal length);
    # 1. Ширина наружной доли околоцветника(sepal width);
    # 1. Длина внутренней доли околоцветника(petal length);
    # 1. Ширина внутренней доли околоцветника(petal width).
iris=datasets.load_iris()   
```

```python
print(iris)
```

```python
# Целевая переменная - это сорт Ириса:
# 0 - Setosa, 1 - Versicolor, 2 - Virginica.
X = iris.data
Y = iris.target
```

```python
k_means3 = KMeans(n_clusters = 3).fit(X)
```

```python
print(k_means3.cluster_centers_)
```

```python
fig = plt.figure(figsize = (10, 4))
ax = fig.add_subplot(121)
plt.scatter(X[:, 0], X[:, 1], c = k_means3.labels_ , marker = 'o', cmap = 'rainbow')
ax = fig.add_subplot(122)
plt.scatter(X[:, 0], X[:, 1], c = Y, marker = 'o', cmap = 'rainbow')
```

```python
fig = plt.figure(figsize = (10, 4))
ax = fig.add_subplot(121)
plt.scatter(X[:, 0], X[:, 2], c = k_means3.labels_ , marker = 'o', cmap = 'rainbow')
ax = fig.add_subplot(122)
plt.scatter(X[:, 0], X[:, 2], c = Y, marker = 'o', cmap = 'rainbow')
```
