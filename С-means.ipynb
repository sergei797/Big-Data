```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.cluster import KMeans
%pip install fuzzy-c-means 
from fcmeans import FCM
```

```python
X = np.array([[0, 0],
              [1, 0],
              [0, 1],
              [1, 1],
              [2, 1],
              [1, 2],
              [2, 2],
              [3, 2],
              [6, 6],
              [7, 6],
              [8, 6],
              [6, 7],
              [7, 7],
              [8, 7],
              [9, 7],
              [7, 8],
              [8, 8],
              [9, 8],
              [8, 9],
              [9, 9],])
```

```python
fcmnew = FCM(n_clusters = 2).fit(X)
```

```python
fcm = FCM(n_clusters = 2)
fcm.fit(X)
```

```python
# outputs
fcm_centers = fcm.centers
fcm_labels = fcm.predict(X)
#plot result
f, axes = plt.subplots(1, 2, figsize = (11, 5))
axes[0].scatter(X[:, 0], X[:, 1])
axes[1].scatter(X[:, 0], X[:, 1], c=fcm_labels)
axes[1].scatter(fcm_centers[:, 0], fcm_centers[:, 1], marker = '+', s = 200, c = 'b')
plt.show()
```

```python
# произвольное множество точек
n_samples = 3000
X = np.concatenate((
    np.random.normal((-2, -2), size = (n_samples, 2)),
    np.random.normal((2, 2),  size = (n_samples, 2))
))
X
```

```python
fcm = FCM(n_clusters = 2)
fcm.fit(X)
```

```python
# метод нечеткой кластеризации
# outputs
fcm_centers = fcm.centers 
fcm_labels = fcm.predict(X)

# plot result
f, axes = plt.subplots(1, 2, figsize = (11, 5))
axes[0].scatter(X[:, 0], X[:, 1])
axes[1].scatter(X[:, 0], X[:, 1], c=fcm_labels)
axes[1].scatter(fcm_centers[:, 0], fcm_centers[:, 1], marker = '+', s = 200, c = 'b')
plt.show()
```

```python
%pip install scikit-fuzzy
from _future_ import division, print_function
import skfuzzy as fuzz
colors = ['b', 'orange', 'g','r', 'c', 'm', 'y', 'k', 'Brown','ForestGreen']
# Define three cluster centers
centers = [[4, 2],
           [1, 7],
           [5, 6]]
# Define three cluster sigmas in x and y, respectively
sigmas = [[0.8, 0.3],
          [0.3, 0.5],
          [1.1, 0.7]]
# Generate test data
np.random.seed(42) #Set seed for reproducibility
xpts = np.zeros(1)
ypts = np.zeros(1)
labels = np.zeros(1)
for i, ((xmu,ymu), (xsigma, ysigma)) in enumerate(zip(senters, sigmas)):
    xpts = np.hstack((xpts, np.random.standart_normal(200) * xsigma + xmu))
    ypts = np.hstack((ypts, np.random.standart_normal(200) * xsigma + ymu))
    labels = np.hstack((labels, np.ones(200) * i))
# Visualise the test data
fig0, ax0 = plt.subplots()
for label in range(3):
    ax0.plot(xpts[labels == label], ypts[labels == label],'.',
             color = colors[label])
ax0.set_title('Test data: 200 points x3 clusters.')    
```
