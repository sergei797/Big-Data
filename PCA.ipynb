```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
# скачивание данных с ирисами фишера
# 1.длина наружной доли околоцветника(sepal length);
# 2.ширина наружной доли околоцветника(sepal windth);
# 3.длина внутренней доли околоцветника(petal windth);
# 4.ширина внутренней доли околоцветника(petal windth).
iris = datasets.load_iris()
```

```python
print(iris)
```

```python
# целевая переменная - это сорт Ириса:
# 0 - Setosa, 1 - Versicolor, 2 - Virginica.
X = iris.data
Y = iris.target
```

```python
X
```

```python
pca = decomposition.PCA()
pca.fit(X)
PCA()
```

```python
# вычисляем дисперсии
explained_variance = pca.explained_variance_
explained_variance
```

```python
# критерий информативности.Вычисляем относительную долю суммарной дисперсии
# обусловленную одним, двумя, тремя, и четырьмя главными компонентами
explained_variance_ratio = pca.explained_variance_ratio_
explained_variance_ratio
```

```python
# сингулярные собственные значения
print(pca.singular_values_)
```

```python
# проверяем с помощью(приведенной ранее теории)numpy получение результата
X = np.transpose(X)
# вычисляем ковариационную матрицу
covmat = np.cov(X)
covmat
```

```python
# находим собственные значения и собственные векторы ковариационной матрицы
A = np.linalg.eig(covmat)
A
```

```python
# отделение собственных векторов
Eigvect = A[1]
Eigvect
```

```python
# выбираем первую и вторую главные компоненты
MainComp = np.transpose(Eigvect[:,0:2])
MainComp
```

```python
# строим массив данных для кластеризации
X_new = np.dot(MainComp, X)
X_new = np.transpose(X_new)
X_new
```

```python
# два варианта кластеризации
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
```

```python
k_means = KMeans(n_clusters = 3).fit(X_new)
```

```python
# визуализация
fig = plt.figure(figsize=(10, 4))
ax = fig.add_subplot(121)
plt.scatter(X_new[:, 0], X_new[:,  1], c = k_means.labels_, marker = 'o', cmap = 'rainbow')
ax = fig.add_subplot(122)
plt.scatter(X_new[:, 0], X_new[:, 1], c = Y, marker = 'o', cmap = 'rainbow')
```

```python
# иерархическая кластеризация
AG_clustering = AgglomerativeClustering(n_clusters = 3, linkage = 'ward').fit(X_new)
fig = plt.figure(figsize=(10, 4))
ax = fig.add_subplot(121)
plt.scatter(X_new[:, 0], X_new[:,  1], c = k_means.labels_, marker = 'o', cmap = 'rainbow')
ax = fig.add_subplot(122)
plt.scatter(X_new[:, 0], X_new[:, 1], c = Y, marker = 'o', cmap = 'rainbow')
```

```python

from collections import OrderedDict
from functools import partial
from time import time

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn import manifold, datasets

Axes3D

n_points = 2000 
X, color = datasets.make_s_curve(n_points, random_state = 10)
# n_neighbors = 10
n_components = 2
# Create figure
fig = plt.figure(figsize = (30, 15))
fig.suptitle("Manifold Leaning (t - SNE) with %i points"
             %(2000,), fontsize = 14)
# Add 3D scatter plot
ax = fig.add_subplot(251, projection = '3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c = color, cmap=plt.cm.Spectral)
ax.view_init(4, -72)
```

```python
# Set-up manifold method
methods = OrderedDict()

methods['t - SNE perplecxity = 5'] = manifold.TSNE(n_components = n_components, 
                                      random_state = 0, perplexity = 5)
methods['t - SNE perplecxity = 15'] = manifold.TSNE(n_components = n_components, 
                                      random_state = 0, perplexity = 15)
methods['t - SNE perplecxity = 30'] = manifold.TSNE(n_components = n_components, 
                                      random_state = 0, perplexity = 30)
methods['t - SNE perplecxity = 50'] = manifold.TSNE(n_components = n_components, 
                                      random_state = 0, perplexity = 50)
```

```python
# Plot results
for i, (label, method) in enumerate(methods.items()):
    t0 = time()
    Y = method.fit_transform(X)
    t1 = time()
    print("%s (%.2g sec)" % (label, t1 - t0))
    ax = fig.add_subplot(2, 5, 2 + i + (i > 3))
    ax.scatter(Y[:, 0], Y[:, 1], c = color, cmap=plt.cm.Spectral)
    ax.set_title("%s (%.2g sec)" %  (label, t1 - t0))
    ax.axis('tight')
    plt.show
```
